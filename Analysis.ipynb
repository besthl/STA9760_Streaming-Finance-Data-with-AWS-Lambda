```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock = pd.read_csv('Desktop/2020 Spring/CIS 9760 Big data techlogy/project_3/results.csv')
```

```python
print(stock.head())
```

## Daily Highest Price for each company

```python
highest = stock.groupby('name').max()['high'].to_frame()
print(highest.head())
```

```python
plt.bar(highest.index, highest.high,color='skyblue')
plt.title('Daily Highest Price for each company')
plt.xlabel('company name')
plt.ylabel('stock price')
plt.show()
```

## Average Daily Stock price for each company

```python
avg_stock = stock.groupby('name').mean()['high'].to_frame()
plt.plot(avg_stock,color='purple')
plt.title('Average Daily Stock Price for each company')
plt.xlabel('company name')
plt.ylabel('stock price')
plt.show()
```
